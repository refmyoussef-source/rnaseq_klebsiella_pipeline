```python
!echo "===> Creating directory for raw FASTQ files..."
!mkdir -p 01_fastq_files
```

```python
%%bash
# "هادي "علامة سحرية" (magic command) كتقول ل Jupyter: "هاد الخانة كلها غتكون Bash"
# (ملاحظة: خاصها تكون هي "السطر" (line) "الأول" (first) ف "الخانة" (cell))

echo "===> Converting SRA to FASTQ (9 samples)... This will take a long time."

# "اللائحة" (list) ديال "الأرقام" (numbers) لي "نزلنا" (downloaded)
SRR_IDS="SRR21159796 SRR21159797 SRR21159798 SRR21159793 SRR21159794 SRR21159795 SRR21159790 SRR21159791 SRR21159792"

# "الحلقة" (loop) (دابا غتخدم 100% حيت "المتغير" (variable) ف نفس الخانة)
for srr in $SRR_IDS; do
    echo "===> Processing $srr ..."

    # "الأمر" (command) "الحقيقي" (real) ديال "التحويل" (conversion)
    fasterq-dump \
        --split-files \
        --threads 4 \
        -O 01_fastq_files \
        -p \
        00_sra_data/$srr/$srr.sra

    echo "===> Finished $srr"
done

echo "===> FASTQ conversion complete."
```

```python
!echo "===> Installing QC Tools (FastQC & MultiQC)..."
!mamba install -c bioconda fastqc multiqc -y
```

```python
%%bash

echo "===> Re-Running FastQC on all 18 FASTQ files..."
echo "This will take some time..."

# -o [المجلد فين غتخرج النتائج] [الملفات لي غنحللو]
fastqc -o 01_fastqc_raw 01_fastq_files/*.fastq

echo "===> FastQC complete."
```

```python
%%bash

echo "===> Running MultiQC to summarize all reports..."

# -o [المجلد فين غتخرج النتائج] [المجلد لي فيه النتائج ديال FastQC]
multiqc -o 01_fastqc_raw 01_fastqc_raw

echo "===> MultiQC complete."
echo "===> Check the report: 01_fastqc_raw/multiqc_report.html"
```
