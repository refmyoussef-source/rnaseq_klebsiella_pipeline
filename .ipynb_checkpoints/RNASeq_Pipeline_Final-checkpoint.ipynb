# Staphylococcus aureus (PRJNA872000) RNA-Seq Analysis Pipeline

This notebook documents the complete bioinformatic workflow and differential gene expression (DGE) analysis for PRJNA872000, investigating antibiotic resistance mechanisms in *Staphylococcus aureus*.

The analysis compares 3 experimental groups (n=3 replicates each):
1.  **control**: Wild Type (Untreated Control)
2.  **treated_NO**: Nitric Oxide (NO) treated group.
3.  **treated_H2O2**: Hydrogen Peroxide (H2O2) treated group.

---
### **Part 1: Bioinformatic Data Processing (Workflow Summary)**
*This section summarizes the heavy processing steps performed on the server. The code is provided for documentation.*

1.  **Data Retrieval:** Raw sequencing reads (`.sra`) were downloaded from the NCBI SRA database using the `sratoolkit`.
2.  **Quality Control (Raw):** Initial quality was assessed using `FastQC` & `MultiQC`.
3.  **Read Trimming:** Adapters and low-quality bases were removed using `Trimmomatic`.
4.  **Quality Control (Trimmed):** Post-trimming quality was verified with `FastQC` & `MultiQC`.
5.  **Genome Indexing:** A reference genome index was built for *S. aureus* using `HISAT2`.
6.  **Read Alignment:** Trimmed reads were aligned to the reference genome using `HISAT2`, producing `.bam` files.
7.  **Read Counting:** Aligned reads were counted against gene features (`.gtf`) using `featureCounts` to generate a raw counts matrix.
8.  **DGE Analysis:** The counts matrix was analyzed using `DESeq2` in R to identify statistically significant differentially expressed genes (DEGs).

---
### **Part 2: Data Visualization & Interpretation (Live Analysis)**
*This section uses Python to load the final statistical results (`.csv` files generated by DESeq2) and visualize the findings.*

```python
# --- Part 2: Live Analysis Setup ---
# Import necessary Python libraries for data analysis and visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set visualization style
sns.set_style("whitegrid")

print("Libraries imported successfully:")
print(f"- pandas version: {pd.__version__}")
print(f"- seaborn version: {sns.__version__}")
```

```python
# Define file paths for the DESeq2 results
file_h2o2_results = "07_deseq2_results/H2O2_vs_Control_results.csv"
file_no_results = "07_deseq2_results/NO_vs_Control_results.csv"

# Load the CSV files into pandas DataFrames
# We set the first column (Unnamed: 0) as the index, which contains the GeneIDs
try:
    res_h2o2_df = pd.read_csv(file_h2o2_results, index_col=0)
    res_no_df = pd.read_csv(file_no_results, index_col=0)

    print("Successfully loaded DESeq2 results.")

    # Display the first 5 rows of the H2O2 vs Control results
    print("\n--- Head of H2O2 vs Control DataFrame ---")
    print(res_h2o2_df.head())

except FileNotFoundError as e:
    print(f"ERROR: File not found. Make sure the path is correct.")
    print(e)
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# --- Define Significance Criteria ---
# We are looking for genes that are statistically significant
# (padj < 0.05 is a common threshold)
P_ADJ_THRESHOLD = 0.05

# (Optional) We also often filter for genes that changed by a certain amount
# (e.g., log2FoldChange > 1 means a 2-fold increase)
L2FC_THRESHOLD = 1.0 

# --- Clean and Filter Data ---

# 1. Clean the H2O2 results
# We must remove 'NaN' rows first, otherwise the filter will fail
h2o2_cleaned_df = res_h2o2_df.dropna()

# 2. Clean the NO results
no_cleaned_df = res_no_df.dropna()

# 3. Apply the filters to find significant genes

# Filter for H2O2
sig_h2o2_df = h2o2_cleaned_df[
    (h2o2_cleaned_df['padj'] < P_ADJ_THRESHOLD) &
    (abs(h2o2_cleaned_df['log2FoldChange']) >= L2FC_THRESHOLD)
]

# Filter for NO
sig_no_df = no_cleaned_df[
    (no_cleaned_df['padj'] < P_ADJ_THRESHOLD) &
    (abs(no_cleaned_df['log2FoldChange']) >= L2FC_THRESHOLD)
]


# --- Print Summary ---
print(f"--- Analysis Summary ---")
print(f"Total genes analyzed (H2O2): {len(h2o2_cleaned_df)}")
print(f"Total genes analyzed (NO):   {len(no_cleaned_df)}")
print(f"\nFound {len(sig_h2o2_df)} significant genes in H2O2 vs Control.")
print(f"Found {len(sig_no_df)} significant genes in NO vs Control.")

# Display the top 10 most significant genes for H2O2
# We sort by 'padj' (the p-value) to see the most significant ones first
print("\n--- Top Significant Genes (H2O2 vs Control) ---")
print(sig_h2o2_df.sort_values(by="padj").head(10))
```

```python
# --- Part 3: Visualization (Step 5) ---
# We will now generate the Volcano Plot

print(f"Generating Volcano Plot for H2O2 vs Control...")

# We use the 'h2o2_cleaned_df' which contains the 81 valid genes
df_to_plot = h2o2_cleaned_df.copy()

print(f"(Plotting {len(df_to_plot)} genes that had valid data)")

# --- 1. Prepare data for plotting ---

# Add a column for -log10(pvalue) for the Y-axis
# We add a tiny number (1e-10) to avoid errors when pvalue is 0
df_to_plot['-log10(pvalue)'] = -np.log10(df_to_plot['pvalue'] + 1e-10)

# Add a column to categorize genes (this is for coloring the plot)
df_to_plot['Significance'] = 'Not Significant'

# We define the significance thresholds (same as before)
P_ADJ_THRESHOLD = 0.05
L2FC_THRESHOLD = 1.0

# Color UP-regulated genes (if they pass BOTH thresholds)
# (We already know this will find 0 genes, but this is the correct code)
df_to_plot.loc[
    (df_to_plot['padj'] < P_ADJ_THRESHOLD) & 
    (df_to_plot['log2FoldChange'] >= L2FC_THRESHOLD), 
    'Significance'
] = 'Up-regulated (Significant)'

# Color DOWN-regulated genes (if they pass BOTH thresholds)
df_to_plot.loc[
    (df_to_plot['padj'] < P_ADJ_THRESHOLD) & 
    (df_to_plot['log2FoldChange'] <= -L2FC_THRESHOLD), 
    'Significance'
] = 'Down-regulated (Significant)'


# --- 2. Create the Volcano Plot ---

plt.figure(figsize=(12, 8))

volcano_plot = sns.scatterplot(
    data=df_to_plot,
    x='log2FoldChange',
    y='-log10(pvalue)',
    hue='Significance',  # Color points based on our new column
    palette={ # Define colors
        'Not Significant': 'grey',
        'Up-regulated (Significant)': 'red',
        'Down-regulated (Significant)': 'blue'
    },
    s=20, # size of points (small)
    alpha=0.6 # transparency
)

# --- 3. Add labels and threshold lines ---

plt.title('Volcano Plot: H2O2 vs Control', fontsize=16)
plt.xlabel('Log2 Fold Change (L2FC)', fontsize=12)
plt.ylabel('-log10(p-value)', fontsize=12)

# Add the p-value threshold line (horizontal line)
# This line shows the border of statistical significance (p=0.05)
p_thresh = -np.log10(0.05) 
plt.axhline(y=p_thresh, color='red', linestyle='--', linewidth=1)

# Add the Log2 Fold Change threshold lines (vertical lines)
plt.axvline(x=L2FC_THRESHOLD, color='red', linestyle='--', linewidth=1)
plt.axvline(x=-L2FC_THRESHOLD, color='red', linestyle='--', linewidth=1)

plt.legend(title='Gene Status')

# Display the plot
plt.show()
```
